## 학습데이터 로드

```python
data_dir = Path('./data/')
trn_file = data_dir / 'train.csv'
seed = 42
```

```python
trn = pd.read_csv(trn_file, index_col=0)
print(trn.shape)
trn.head()
```

```python
sns.pairplot(trn.sample(n=10000), vars=['u', 'dered_u'], hue='class', size=5)
```

```python
# class 가 1, 2 인 데이터로 진행
X = trn[trn['class'] != 0][['u']].values
y = trn[trn['class'] != 0]['dered_u'].values
print(X.shape, y.shape)
```

## RMSE 손실함수 정의

```python
rmse = lambda y, p: np.sqrt(np.mean((y - p) ** 2))
```

## NumPy를 이용한 선형회귀

```python
alpha = .001
theta0 = .1
theta1 = .1
epoch = 4
```

```python
x = X.flatten()
print(x.shape)
```

```python
rmses = []
for i in tqdm(range(epoch)):
    for x_i, y_i in zip(x, y):
        p_i =  theta0 + theta1 * x_i
        theta0 = theta0 + alpha * (y_i - p_i)
        theta1 = theta1 + alpha * (y_i - p_i) * x_i
    rmses.append(rmse(y, theta0 + theta1 * x))
```

```python
theta1, theta0
```

```python
rmses
```

```python
plt.plot(rmses)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X, y, alpha=.1)
plt.plot(X, theta0 + theta1 * x, color='darkorange', linewidth=.5)
plt.xlim(0, 40)
plt.ylim(0, 40)
plt.xlabel('u')
plt.ylabel('dered_u')
```

## scikit-learn을 이용한 선형회귀

```python
from sklearn.linear_model import LinearRegression

X = trn[trn['class'] != 0].drop('class', axis=1)
y = trn[trn['class'] != 0]['class'].values - 1

lr = LinearRegression()
lr.fit(X, y)
p = lr.predict(X)
```

```python
# Reduced Mean Loss function
# Reduced Mean 손실함수
rmse(y, p)
```

```python
# coef_ : Array of shape
# intercept_ : float or array of shape
lr.coef_, lr.intercept_
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X, y, alpha=.1)
plt.plot(X, p, color='darkorange', linewidth=.5)
plt.xlim(0, 40)
plt.ylim(0, 40)
plt.xlabel('u')
plt.ylabel('dered_u')
```

## scikit-learn을 이용한 로지스틱회귀

```python
X = trn[trn['class'] != 0].drop('class', axis=1)
y = trn[trn['class'] != 0]['class'].values - 1
```

```python
np.unique(y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss, roc_auc_score

lr = LogisticRegression()
lr.fit(X, y)
p = lr.predict_proba(X)[:, 1]
```

```python
lr.coef_, lr.intercept_
```

```python
# Log 손실함수
log_loss(y, p)
```

```python
# Logistic Regression Accuracy
roc_auc_score(y, p)
```

```python
# from kaggler.metrics import plot_roc_curve, plot_pr_curve
from sklearn.metrics import roc_curve, precision_recall_curve
```

```python
def plot_roc_curve(y, p):
    fpr, tpr, _ = roc_curve(y, p)

    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')


def plot_pr_curve(y, p): 
    precision, recall, _ = precision_recall_curve(y, p)

    plt.step(recall, precision, color='b', alpha=0.2, where='post')
    plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
```

```python
plt.figure(figsize=(16, 16))
plt.subplot(1, 2, 1)
plot_roc_curve(y, p);
plt.title('AUC Curve')
plt.subplot(1, 2, 2)
plot_pr_curve(y, p);
plt.title('Precision-Recall Curve')
```
